```python
"""<선형(연결리스트)>
주제: 플레이리스트 프로그램
요구사항: 
1) 입력: 노래 제목, 가수, 장르, 메모, 관심 표시를 입력받는다. (장르는 번호로 선택하도록 만든다. / 관심 표시는 하트 최대 5개로 한다.)
2) 출력: 하트 개수 순서대로 노래들을 출력한다. 
3) 기능들: 정보 수정 기능(관심표시나 가수 같은 정보 수정), 노래 추가, 노래 삭제
"""
```

```python
class Node:
    def __init__(self, data): #data는 list()
        self.song = data[0]
        self.singer=data[1]
        self.genre=data[2]
        self.memo=data[3]
        self.heart=data[4]
        self.next = None
```

```python
class PlayList:
    def __init__(self):
        self.head = None # head node
    
    #헤더부터 탐색해 뒤에 새로운 노드 추가하기
    def append(self, data):
        cur = self.head
        if cur is None: #처음이면
            cur = self.head
            self.head = Node(data)
        else:
            # 꼬리 노드 탐색
            while cur.next is not None:
                cur = cur.next

            cur.next = Node(data)
    
    #제목으로 노래 제거
    def remove(self, name):
        if self.head is not None:
            if self.head.song == name:
                if self.head is not None:
                    self.head = self.head.next
                return True

            cur = self.head
            while cur.next is not None:
                if cur.next.song == name:
                    cur.next = cur.next.next
                    return True
                cur = cur.next
        return False
    
    #하트 개수로 노래 찾기
    def search_by_heart(self, n):
        cur = self.head
        while cur is not None:
            if cur.heart == n:
                return cur
            cur = cur.next
        return -1
    
    #Output : 하트 개수 순으로 노래 출력
    def printPlayList(self):
        cur = self.head
        if cur==None:
            print("곡이 없습니다.")
            return
        
        heart_count=[]
        while cur is not None:
            heart_count.append(cur)
            cur = cur.next
        heart_count.sort(key=lambda a:a.heart, reverse=True)
        # heart_count.sort(reverse=True)
        # for i in heart_count:
        #     cur=self.search_by_heart(i)
        #     print(f"❤:{i} {cur.song} - {cur.singer}")
        for cur in heart_count:
            print(f"❤:{cur.heart} [{cur.song}] - {cur.singer} , {cur.genre}, {cur.memo}")
    
    # 가수 정보 변경
    def edit_singer(self, song, new_singer):
        cur = self.head
        while cur is not None:
            if cur.song == song:
                cur.singer=new_singer
                print(f"{cur.song}의 가수 정보가 {new_singer}로 변경되었습니다.")
                return
            cur = cur.next
        print("곡의 정보를 찾을 수 없습니다.")
    
    # 하트 개수 변경   
    def edit_heart(self, song, new_heart):
        cur = self.head
        while cur is not None:
            if cur.song == song:
                cur.heart=new_heart
                print(f"{cur.song}의 하트 개수가 {new_heart}개로 변경되었습니다.")
                return
            cur = cur.next
        print("곡의 정보를 찾을 수 없습니다.")    
        
```

```python
if __name__=="__main__":
    # PlayList 생성
    pl=PlayList()
    
    #Start Menu
    while True:
        data=[]
        cmd=input("1. 곡추가 2. 곡삭제 3. 가수 정보 수정 4. 관심표시 수정 5. 프로그램 종료")
        
        # Input : 곡 추가
        if cmd=='1':
            song=input("노래 제목 입력 >> ")
            singer=input("가수 입력 >> ")
            genre=input("장르 입력 >> ")
            memo=input("메모 입력 >> ")
            heart=input("하트 개수 입력 (1~5 사이 값) >> ")
            
            data.append(song)
            data.append(singer)
            data.append(genre)
            data.append(memo)
            data.append(heart)
        
            pl.append(data)
        
        # 2. 곡 삭제
        elif cmd=='2':
            print('노래 삭제')
            song=input("삭제하려는 노래의 제목 입력 >> ")
            pl.remove(song)
        
        # 3. 가수정보 수정
        elif cmd=='3':
            print("가수 수정")
            song=input("가수 수정하려고 하는 노래의 제목 입력 >> ")
            new_singer= input("새로운 가수 이름 입력 >> ")
            pl.edit_singer(song, new_singer)
            
        # 4. 하트개수 수정
        elif cmd=='4':
            song=input("관심표시 수정하려고 하는 노래의 제목 입력 >> ")
            new_heart= input("새로운 하트 개수 입력 >> ")
            pl.edit_heart(song, new_heart)
            
        # 5. 프로그램 종료
        elif cmd=='5':
            print("프로그램이 종료됩니다.")
            break
        
        # 예외처리
        else: 
            print("==============잘못된 입력값입니다.==============")
            
        # Output : 하트 개수대로 곡 정보들 Output 출력
        print('=='*23)
        print("\t\t♫<Play List>♫")
        pl.printPlayList()
        print('=='*23)
        print()
        
        
```
